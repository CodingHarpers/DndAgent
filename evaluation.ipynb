```python
import pandas as pd
```

```python
file_path = "backend/data/logs/evaluation.xlsx"
df = pd.read_excel(file_path)
```

```python
df.describe()
```

```python
gemini_file_path = "backend/data/logs/gemini.xlsx"
gemini_df = pd.read_excel(gemini_file_path)
gemini_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))

ax = sns.countplot(x='gpt_score', data=df, palette='viridis', hue='gpt_score', legend=False)

plt.title('Distribution of ARCANA Performance Scores', fontsize=16, pad=20)
plt.xlabel('Score (1-5)', fontsize=12)
plt.ylabel('Count of Interactions', fontsize=12)

for p in ax.patches:
    height = int(p.get_height())
    ax.text(p.get_x() + p.get_width() / 2., height + 0.5, 
            f'{height}', 
            ha="center", va="bottom", fontsize=11, fontweight='bold')

plt.ylim(0, df['gpt_score'].value_counts().max() * 1.15)

plt.show()
```
